### Data Scraping

The data I gathered was aquired from UNICEF's Multiple Indicator Cluster's Survey. This survey excompasses various sets of data that track progress of women's health and development in middle and lower income countries. The data that I specifically chose was the "Attitudes towards Wife Beating" data set that tracked both men's and women's perceptions towards wife beating.The data collected by UNICEF followed a multistage clustering technique that encompessed about 5,000-40,000 households intra-nationally. Attitudes towards domestic abuse were measured by 5 indicators: "(1) she goes out without telling him, (2) she neglects the children, (3) she argues with him, (4) she refuses to have sex with him, (5) she burns the food." These questions were replicated for both men and women and remained constant if not slightly tweaked internationally to match national sentiments of women's roles within the country.

The first set of data I collected was in regards to male sentiment about hitting women in regards to these conditions:

```python
#Scraping data about male sentiment in regards to domestic abuse

#!pip install folium
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

#Read in excel file of the data set and then store in data frame
men = pd.read_excel ('Attitudes-towards-wife-beating-Men-database_Nov-2017.xlsx')
men= men.iloc[9:207]

#Clean out any uncessary data and all rows with missing values
#This is an adequate way to handle this form of missing data as the countries that were missing 
#were not countries within the lower and middle class spectrum of countries. 
men.drop(men.columns[[1,3,5,7,9,11,13,15,16]],axis=1,inplace=True)
men = men.replace(r'-', np.nan, regex=True)
men = men.dropna()

#Reset indices to match up with proper row
men = men.reset_index()
men.reindex(index=range(0,73))

#Rename columns
men.rename(columns={'index': 'Country','GLOBAL DATABASES':'Total %','Unnamed: 2':'Urban %','Unnamed: 4':'Rural %', 
                      'Unnamed: 6':'Poorest %','Unnamed: 8':'Second %', 'Unnamed: 10':'Middle %',
                      'Unnamed: 12':'Fourth %', 'Unnamed: 14':'Richest %'}, inplace=True)

men.sample(20)
```

The next set of data I parsed was female sentiment toward's domestic abuse. To parse the data I followed the same procedure as the initial data set but filtered out any additional countries specific to this data set. This way both the male and female data sets contained the same countries for adequate analysis later. 

```python
#Scraping for data in relationship to women's sentiments
women = pd.read_excel ('Attitudes-towards-wife-beating-Women-database_Nov-2017.xls')

#Clean table to remove any unecessary columns and rows
women= women.iloc[9:207]
women.drop(women.columns[[1,3,5,7,9,11,13,15,16]],axis=1,inplace=True)
women = women.replace(r'-', np.nan, regex=True)
women = women.dropna()

#Reset index and recolumn
women = women.reset_index()
women.reindex(index=range(0,73))
women.rename(columns={'index': 'Country','GLOBAL DATABASES':'Total %','Unnamed: 2':'Urban %','Unnamed: 4':'Rural %', 
                      'Unnamed: 6':'Poorest %','Unnamed: 8':'Second %', 'Unnamed: 10':'Middle %',
                      'Unnamed: 12':'Fourth %', 'Unnamed: 14':'Richest %'}, inplace=True)

#FIlter out countries that are not present in the male dataset 
filter1 = women["Country"].isin(men["Country"]) 
women2=women[filter1]

#Reset the index again and display data
women2 = women.reset_index()
del women2['index']
women2.sample(20)
```

### Visualizing Data

Before analyzing the data, I wanted to see the distribution of positive perceptions of domestic abuse across the world. Understanding in what regions these sentiments were concentrated would allow a further undestanding of what kind's of demographics lead to these perceptions and what kind's of social situations perpetrate domestic abuse at large.

```python
#Chorpleth display of data to see concentration of postive sentiments globally
#USed folium for map visualization
import folium
from folium.map import *

#COnstruct general map layout
m = folium.Map(location=[25, 40], zoom_start=3)

#To construct choropleth, needed a display of world breakdown by country to easily populate map
country_geo = 'world-countries.json'

#Read in data into the map and had the color get darker green as perceptions increased
folium.Choropleth(
    geo_data=country_geo,
    name='choropleth',
    data=women,
    columns=['Country', 'Total %'],
    key_on='feature.properties.name',
    fill_color='YlGn',
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name='Approval of Domestic Abuse (%)'
).add_to(m)
import os
folium.LayerControl().add_to(m)
m.save(os.path.join('map.html'))
m
```

From this data we can see how a majority of those accepting of physical violence towards women, falls with in the South Asian, Middle East, and Sub-Saharan African regions while lowest in eastern eurpoean and American regions. This data is reflective of perceptions towards women and their role within society. To gather further analysis, I decided to expand in a deeper analysis of exonomic and social distributions of households and that effect on thes perceptions. 

### Data Analysis and Visualization

To begin, I wanted to see the difference between female and male perceptions of physical abuse towards women. With the understanding that these were male imposed afflictions, I arrived with the perception, that women would have significantly lower justification ofdomestic abuse. I also wanted to break down the analysis based off of demographic to understand the kinds of social and economic factors that implicate these perceptions

```python
#Retrieved the mean value for men that justified physical violence against women for both rural and urban areas
mean_men_urban = men['Urban %'].mean()
mean_men_rural = men['Rural %'].mean()

#Retrieved the mean value for women that justified physical violence against men for both rural and urban areas
mean_women_urban = women['Urban %'].mean()
mean_women_rural = women['Rural %'].mean()

#STored mean values in array for both men and women to display in bar graph
men_mean = [ mean_men_urban, mean_men_rural]
women_mean = [mean_women_urban, mean_women_rural]

#Arranged bar graph and display
barWidth = 0.25
r1 = np.arange(len(men_mean))
r2 = [x + barWidth for x in r1]
fig = plt.figure(figsize=(14,10))
plt.bar(r1, men_mean, color='cadetblue', width=barWidth, edgecolor='white', label='men')
plt.bar(r2, women_mean, color='indianred', width=barWidth, edgecolor='white', label='women')

#Labeled axis and title
plt.title('Average Percentage of Domestic Abuse Justification (Urban vs Rural)', fontweight='bold')
plt.ylabel('Avergae Percentage of Approval', fontweight='bold')
plt.xlabel('Demographic', fontweight='bold')
plt.xticks([r + barWidth for r in range(len(men_mean))], ['Urban %', 'Rural %'])

#DIsplayed plot
plt.legend()
plt.show()



```

From here we can see how women significantly justify physical afflictions from their partner far more than men justify afflicting them upon women. What's also interesting here is how there is a significant gap between urbal and rurl demographics with rural demographics, significantly justfying these actions. 

To expand this even more, I decided to take a look at how perceptions differ across different income levels:


```python
#Identified mean perceptions of men among different social classes
mean_men_poorest = men['Poorest %'].mean()
mean_men_second = men['Second %'].mean()
mean_men_middle = men['Middle %'].mean()
mean_men_fourth = men['Fourth %'].mean()
mean_men_richest = men['Richest %'].mean()

#Identified mean perceptions of women among different social classes
mean_women_poorest = women['Poorest %'].mean()
mean_women_second = women['Second %'].mean()
mean_women_middle = women['Middle %'].mean()
mean_women_fourth = women['Fourth %'].mean()
mean_women_richest = women['Richest %'].mean()


#Stored all data in two arrays indicitive of gender
women_mean = [ mean_women_poorest, mean_women_second,mean_women_middle,mean_women_fourth,mean_women_richest]
men_mean = [mean_men_poorest, mean_men_second,mean_men_middle,mean_men_fourth,mean_men_richest]

barWidth = 0.25

#Fromatted data into bar plot
r1 = np.arange(len(men_mean))
r2 = [x + barWidth for x in r1]
fig = plt.figure(figsize=(15,10))
plt.bar(r1, men_mean, color='cadetblue', width=barWidth, edgecolor='white', label='men')
plt.bar(r2, women_mean, color='indianred', width=barWidth, edgecolor='white', label='women')

#Labeled axis and title
plt.title('Average Percentage of Domestic Abuse Approval per Class Status', fontweight='bold')
plt.ylabel('Avergae Percentage of Approval', fontweight='bold')
plt.xlabel('Class Status', fontweight='bold')
plt.xticks([r + barWidth for r in range(len(men_mean))], ['Poorest %', 'Second %','Middle %', 'Fourth %', "Richest %" ])

#Displayed plot
plt.legend()
plt.show()
```

From here, I noticed a significant jump between perceptions in the highest and lowest income. There was a steady ddecline in justification of physical afflications against women as income class increased. After analyzing both his bar graph as well as the previous ar graph, I noticed a significant trend that I wanted to explore more. Individuals of the upper class and urban regions generally had a higher education level. After being really shocked by extremely high level of justification of physical abuse against women by women, I realized an important factor that may contribute to this is the education levels of men vs women within these lower and middle class countries.To explore this even more, I decided to analyze how education levels play a role in the divide between male and female perceptions of domestic abuse.

### Education Level of Women and Men 

The following data set is also from UNICEF and takes into account for adult literacy. While I was intially considering looking at either primary or secondary education levels, literacy levels seemed to be the best indicator of basic and essential education and provide for easier comparison across data sets. 

```python
#Scrape data of adult literacy and store in data frame
literacyRates = pd.read_excel ('Table-Youth-and-Adult-Literacy-Rate-updated-Oct.-2015_78.xlsx', sheet_name='Adult literacy rate')

#Remove and clean unecessary data and reset index accordingly
literacyRates= literacyRates.iloc[10:207]
literacyRates.drop(literacyRates.columns[[0,2,4,6,7]],axis=1,inplace=True)
literacyRates = literacyRates.replace(r'-', np.nan, regex=True)
literacyRates = literacyRates.dropna()
literacyRates = literacyRates.reset_index()
literacyRates.reindex(index=range(0,73))
literacyRates.drop(literacyRates.columns[[0]],axis=1,inplace=True)
literacyRates.rename(columns={'level_1': 'Country','Unnamed: 1':'Average %','Unnamed: 3':'Male %', 
                      'Unnamed: 5':'Female %'}, inplace=True)
filter1 = literacyRates["Country"].isin(women["Country"]) 

#Restructe data so it only matches with countries contained within the women data set
literacyRates=literacyRates[filter1]
literacyRates = literacyRates.reset_index()
del literacyRates['index']
literacyRates.sample(20)
```

Merge the literacy data set with both the male and femal data sets of justifications of physical violence

```python
#PErform an outer merge of data tables to keep data maintained per each country 
#This merge uses data from the literacy dat set as well as the women justification data set
temp = literacyRates.merge(women, how = 'outer')
temp=temp.dropna()
temp.head()
```

```python
#PErform an outer merge of data tables to keep data maintained per each country 
#This merge uses data from the literacy dat set as well as the men justification data set
temp2 = literacyRates.merge(men, how = 'outer')
temp2=temp2.dropna()
temp2.head()
```

### Visualization of Literacy Rates and Percentage of Justification

The following marks the literacy rates and justification of physical abuse per each gender. The goal of this data is to see how each varies across each gender. 

```python
#Gather mean data about literacy rates of men within these countries as well as the mean justification
mean_men_literacy = temp2['Male %'].mean()
mean_men_justification = temp2['Total %'].mean()

#Gather the mean literacy rates of women across these countries as well as the averge justification  
mean_women_literacy = temp['Female %'].mean()
mean_women_justification = temp['Total %'].mean()

#Combine avergaes in justification and literacy
#This time data is split by gender and each bar represents a different data set
men_mean = [ mean_men_justification, mean_women_justification]
women_mean = [mean_men_literacy, mean_women_literacy]

#Construct bar plot with given data and display accordingly 
barWidth = 0.25
r1 = np.arange(len(men_mean))
r2 = [x + barWidth for x in r1]
fig = plt.figure(figsize=(14,10))
plt.bar(r1, men_mean, color='firebrick', width=barWidth, edgecolor='white', label='Support Domestic Abuse')
plt.bar(r2, women_mean, color='goldenrod', width=barWidth, edgecolor='white', label='Literacy Rate')

#Label and title axis 
plt.title('Average Percentage of Domestic Abuse Approval Vs Average Literacy Rate', fontweight='bold')
plt.ylabel('Average Percentage', fontweight='bold')
plt.xlabel('Gender', fontweight='bold')
plt.xticks([r + barWidth for r in range(len(men_mean))], ['Men', 'Women'])

#Display plot
plt.legend()
plt.show()
print('Men Literacy: '+ str(mean_men_literacy))
print('Women Literacy: '+ str(mean_women_literacy))
print('Men Justification: ' +  str(mean_men_justification))
print('Women Justification: ' + str(mean_women_justification))
```

From this bar plot we can see how the average literacy rate of women is substantially lower than men. With men at about 76.3% literacy in these lower and middle class countries, it is a significant jump that only about 69% of women are literate. A similar gab exists between womena nd men in terms of justifications of physical abuse with men justifying it at approximately 29.9% and men women at 36.6%. After analyzing this graph it appeared to me that there may in fact be a negative correlation between education and perceptions on domestic violence which is what led me to the following scatter plot. 

```python
#Scatter lot with literacy rate on x axis and percentage of justifation on y-axis
#First scatter plot with female data
x = temp['Female %']
y = temp['Total %']
fig = plt.figure(figsize=(14,10))
plt.scatter(x, y, c='lightsalmon',marker='o',label='women')

#Draw line of best fit for female data set
z = np.polyfit(x, y, 1)
p = np.poly1d(z)
plt.plot(x,p(x),color="indianred", linestyle='dashed')

#Second scatter plot with male data
x2 = temp2['Male %']
y2 = temp2['Total %']
plt.scatter(x2, y2, c='cadetblue', marker="x",label='men')

#Line of best fit for male data
z2 = np.polyfit(x2, y2, 1)
p2 = np.poly1d(z2)
plt.plot(x2,p2(x2),color="teal", linestyle='dashed')

#Labels for axis and title
plt.xlabel("Literacy Rate of Women (%)", fontweight='bold')
plt.ylabel("Women Acceptance of Domestic Abuse (%)", fontweight='bold')
plt.title("Literacy Rate of Women Vs Acceptance of Domestic Abuse", fontweight='bold')
plt.legend(loc='lower left');

#Display plot
plt.show()
```

From this graph, a negative correlation between literacy rate and the percentage of acceptance of physical abuse. An interesting trend here is how the slope of the women data set is significantly steeper than that of the male data set. This goes to show the impact of education and literacy in a women's acceptance of physical abuse.
